# Imports

```python
import time
import sys
import random
import operator
from itertools import product
from joblib import Parallel, delayed, parallel_backend
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import curve_fit
import hnsw
from iris_integration import (
    iris_random,
    iris_with_noise,
    irisint_make_query as make_query,
    irisint_query_to_vector as query_to_vector,
    irisint_distance as distance,
)
```

# Functions

```python
def update_db(db, iris_df, db_size):
    db_current_size, new_samples_lst = db.get_stats()['db_size'], []
    assert (db_size - db_current_size) > 0
    for i in range(db_size - db_current_size):
        random_sample = iris_random()
        query = make_query(random_sample)
        id_random_sample = db.insert(query)
        new_samples_lst.append((id_random_sample, random_sample))
    return db, pd.concat([iris_df, pd.DataFrame(new_samples_lst, columns=["ID", "Template"])], ignore_index=True)
```

```python
def run_single_experiment(i, db, iris_df, noise_level, efSearch, K):
    # np.random.seed(i), random.seed(i)
    random_target_id = random.randint(0, db.get_stats()['db_size']-1)
    noisy_target = iris_with_noise(iris_df.loc[random_target_id, 'Template'], noise_level=noise_level)
    noisy_query = make_query(noisy_target)
    res = db.search(noisy_query, K, ef=efSearch)
    return random_target_id in list(map(operator.itemgetter(1), res))
```

# Configurations

```python
# Constants values
num_experiments = 20
noise_level = 0.3
K = 5
m_L = 0.3
n_jobs = 4 # Fit to CPU
precision_thres = 0.6
```

```python
# Tested values
num_db_elements = np.linspace(1000, 10000, 10, dtype=int) # np.linspace(1000, 50000, 50, dtype=int)
Ms = [64, 128, 256] # np.linspace(2, 256, 128, dtype=int)
efConstructions = [64, 128] # np.linspace(2, 256, 128, dtype=int)
efSearchs = [64, 128] # np.linspace(2, 256, 128, dtype=int)
total_iterations = len(num_db_elements) * len(Ms) * len(efConstructions) * len(efSearchs)
```

# Stats computation

```python
rows = [] # Will eventually include the statistics from all parameter combinations
for s, (M, efConstruction, efSearch) in enumerate(product(Ms, efConstructions, efSearchs)):
    # Initiate HNSW instance
    db = hnsw.HNSW(
        M=M, 
        efConstruction=efConstruction, 
        m_L=m_L, 
        distance_func=distance, 
        query_to_vector_func=query_to_vector
    )
    iris_df = pd.DataFrame()
    
    for j, db_size in enumerate(num_db_elements):   
        # Print progress update
        progress_ratio = (s*len(num_db_elements)+j+1) / total_iterations
        sys.stdout.write(f"\rCurrent progress out of {total_iterations} parameter combinations (with {num_experiments} experiments each): {progress_ratio:.1%}")
        sys.stdout.flush()

        # Update the db (use the insertion feature)
        db, iris_df = update_db(db, iris_df, db_size)

        # Run num_experiments in parallel (to average the precision over multiple random seeds)
        with parallel_backend('threading'): # Reassure with Aurel that Parallel computing isn't problematic (changes the db)
            results = Parallel(n_jobs=n_jobs)(
                delayed(run_single_experiment)(i, db, iris_df, noise_level, efSearch, K) for i in range(num_experiments)
            )

        # Update results
        new_row = [db_size, K, M, efConstruction, efSearch, sum(results) / num_experiments]
        rows.append(new_row)
stats_df = pd.DataFrame(rows, columns=['db_size', 'K', 'M', 'efConstruction', 'efSearch', 'precision'])
stats_df.head()
```
